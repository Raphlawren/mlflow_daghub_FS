### import important libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2, SelectFromModel
from sklearn.linear_model import LogisticRegression, LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import time
```

```python
# standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

## FILTER VS EMBEDDED VS WRAPPER

```python
### FILTER METHOD: Chi-SQUARE

start = time.time()
X_train_abs = np.abs(X_train) ## negative values are bad for chi2
selector = SelectKBest(score_func = chi2, k=30)
X_train_filter = selector.fit_transform(X_train_abs, y_train)
X_test_filter = selector.transform(X_test)
clf_filter = LogisticRegression(max_iter=1000)
clf_filter.fit(X_train_filter, y_train)
y_pred_filter = clf_filter.predict(X_test_filter)
acc_filter =accuracy_score(y_test, y_pred_filter)
time_filter = time.time() - start
print(f"\n[Filter] Accuracy: {acc_filter:.4f} | Features: {X_train_filter.shape[1]} | Time: {time.time() - start:.2f}s")
```

```python
selected_indices = selector.get_support(indices=True)
filtered_selected_feature_names = X.columns[selected_indices].tolist()
print("Selected features (Filter - Chi2):")
print(filtered_selected_feature_names)
```

```python
###  EMBEDED METHOD: LASSO
start = time.time()
lasso = LassoCV(cv=5, max_iter  = 10000)
lasso.fit(X_train, y_train)
model_emb = SelectFromModel(lasso, prefit=True)

X_train_emb = model_emb.transform(X_train)
X_test_emb = model_emb.transform(X_test)
if X_train_emb.shape[1] == 0:
    print("\n[Embedded] LASSO selected 0 features — skip evaluation.")
else:
    clf_emb = LogisticRegression(max_iter=1000)
    clf_emb.fit(X_train_emb, y_train)
    y_pred_emb = clf_emb.predict(X_test_emb)
    acc_emb = accuracy_score(y_test, y_pred_emb)
    time_emb = time.time() - start
    print(f"\n[Embedded] Accuracy: {acc_emb:.4f} | Features: {X_train_emb.shape[1]} | Time: {time.time() - start:.2f}s")
```

```python
selected_indices = model_emb.get_support(indices=True)
embedded_selected_feature_names = X.columns[selected_indices].tolist()
print("Selected features (Embedded - LASSO):")
print(embedded_selected_feature_names)
```

```python
# ! pip install pyswarms
from pyswarms.discrete import BinaryPSO
from sklearn.model_selection import cross_val_score

def fitness_function(mask):
    scores = []
    for particle in mask:
        idx = np.where(particle == 1)[0]
        if len(idx) == 0:
            scores.append(0)  # prevent empty feature set
            continue
        X_sub = X_train[:, idx]
        clf = LogisticRegression(max_iter=1000)
        acc = cross_val_score(clf, X_sub, y_train, cv=3, scoring='accuracy').mean()
        scores.append(acc)
    return np.array(scores) * -1  # minimize

start = time.time()
n_features = X_train.shape[1]
options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9, 'k':3, 'p':2}
optimizer = BinaryPSO(n_particles=10, dimensions=n_features, options=options)
cost, pos = optimizer.optimize(fitness_function, iters=10)
selected_idx = np.where(pos == 1)[0]

if len(selected_idx) == 0:
    print("\n[Wrapper] PSO selected 0 features — skip evaluation.")
else:
    X_train_pso = X_train[:, selected_idx]
    X_test_pso = X_test[:, selected_idx]
    clf_pso = LogisticRegression(max_iter=1000)
    clf_pso.fit(X_train_pso, y_train)
    y_pred_pso = clf_pso.predict(X_test_pso)
    acc_pso = accuracy_score(y_test, y_pred_pso)
    time_pso = time.time() - start
    print(f"\n[Wrapper] Accuracy: {acc_pso:.4f} | Features: {X_train_pso.shape[1]} | Time: {time.time() - start:.2f}s")
```

```python
wrapper_selected_feature_names = X.columns[selected_idx].tolist()
print("Selected features (Wrapper - PSO):")
print(wrapper_selected_feature_names)
```
